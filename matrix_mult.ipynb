```python
import torch
```

### Matrix Multiplication

```python
tensor = torch.tensor([1, 2, 3])
```

```python
# Element wise multiplication
tensor * tensor
```

```python
# Matrix Multiplication
torch.matmul(tensor, tensor) # Inbuilt methods are fast and optimized for faster computations 
```

```python
# Talking transpose
A = torch.tensor([[1, 2],
                  [3, 4],
                  [5, 6]])
B = torch.tensor([[7, 8],
                  [9, 10],
                  [11, 12]])

A @ B
```

```python
A.shape, B.shape
```

```python
B = B.T
```

```python
A @ B
```
