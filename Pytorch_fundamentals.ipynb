## Random Tensors

```python
# Create a random tensor of size (3, 4)
random_tensor = torch.rand(3, 4)
```

```python
random_tensor
```

```python
random_tensor.ndim
```

```python
random_tensor.shape
```

```python
# Create a random tensor with similar shape to an image tensor
random_img_size_tensor = torch.rand(size=(224, 224, 3))
random_img_size_tensor.shape, random_img_size_tensor.ndim
```

## Zeros and Ones

```python
# create a tensor of all zeros 
zeros = torch.zeros(size=(3, 4))
zeros
```

```python
# Create tensors of ones
ones = torch.ones(size=(3, 4))
ones
```

```python
ones.dtype
```

### Creating a range of tensors and tensors-like

```python
# Use torch.range()
one_to_ten = torch.arange(start=1, end=11, step=2)
one_to_ten
```

```python
# Create tensors like
ten_zeros = torch.zeros_like(input=one_to_ten)
ten_zeros
```

## Tensor datatypes

```python
# Float 32 tensor
float_32_tensor = torch.tensor([3.0, 6.0, 9.0], dtype=None,
                            device=None,
                            requires_grad=False)
float_32_tensor
```

```python
float_32_tensor.dtype
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
```

```python
float_16_tensor
```

```python
float_16_tensor * float_32_tensor
```

### Manipulating Tensors (tensor operations)

Tensor Operations include:
* Addition
* Subtraction
* Multiplication (element-wise)
* Division
* Matrix Multiplication

```python
# Creating a tensor
tensor = torch.tensor([1, 2, 3])
tensor + 10
```

```python
# Multiply tensor
tensor * 10
```

```python
tensor - 10
```

```python
# Try out Pytorch inbuilt functions
torch.mul(tensor, 10)
```

```python
torch.add(tensor, 10)
```
